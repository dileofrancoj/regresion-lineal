```python
import numpy as np # lib de matemática
import matplotlib.pyplot as plt # graficar "cositas"
import seaborn as sns # graficas un poco más profesionales
```

```python
from sklearn.datasets import load_boston # dataset que tiene info de precios de casas basados en diferentes features
```

```python
boston = load_boston()
```

```python
print(boston.DESCR)
```

```python
X = boston.data[:,5]
# https://www.kaggle.com/c/boston-housing
```

```python
Y = boston.target
```

```python
sns.scatterplot(x=X, y = Y, alpha=.3)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
X = X.reshape(-1,1)
```

```python
model.fit(X,Y) # lo entreno con los datos historicos
```

```python
print("Pendiente de la recta:", model.coef_)
print("O.O de la recta:", model.intercept_)
```

```python
y_pred = model.predict(X)
```

```python
plt.scatter(X,Y)
plt.plot(X,y_pred, color='red')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(y_pred, Y)
print(mae)
```

```python
mse = mean_squared_error(y_pred, Y)
print(mse)
```

```python
r2 = r2_score(Y,y_pred)
print(r2 * 100)
```

```python
X_2 = np.array([7]).reshape(-1,1)
```

```python
y_pred_2 = model.predict(X_2)
print(y_pred_2)
```
